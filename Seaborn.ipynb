```python
import pandas as pd
df=pd.read_csv('dataset/auto_mpg.csv')
df
```

```python
#1. BOX PLOT
import seaborn as sns
import matplotlib.pyplot as plt
sns.boxplot(data=df,y=df['mpg'])
plt.show()
```

```python
#2. SCATTER PLOT
sns.scatterplot(data=df,x=df['mpg'],y=df['horsepower'])
plt.show()
```

```python
#3. BAR PLOT
sns.barplot(data=df,x=df['model_year'],y=df['horsepower'])
plt.show()
```

```python
#.4. HISTOGRAM
sns.histplot(data=df,x=df['horsepower'])
plt.show()
```

```python
#.5.KDE
sns.histplot(data=df,x=df['horsepower'],kde=True)
```

```python
sns.heatmap(data=df[['mpg']])
plt.show()
```

```python
sns.countplot(data=df,x=df['horsepower'])
plt.show()
```

```python
sns.lineplot(data=df,x=df['horsepower'],y=df['mpg'])
plt.show()
```
