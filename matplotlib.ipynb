```python
#.5. LINE CHART
#XYZ Custom Cars is planning to generate a report comparing the Mileage, 
#Horsepower, and Weight of the cars manufactured in different years, starting from 1970 to 1982.
line_df=df.groupby(by=['model_year'])
line_df=line_df.mean()
 
    
```

```python
##importing the required packages
import matplotlib.pyplot as plt
#creating an empty canvas/figure
fig = plt.figure(figsize=[8,4])
#setting axes
ax = fig.add_axes([0, 0, 1, 1])
#plotting the lines
ax.plot(line_df['mpg']/10,label='Mileage /10')
ax.plot(line_df['weight']/1000,label='Weight /1000')
ax.plot(line_df['horsepower']/100,label='Horsepower /100')
ax.set_title('Mileage vs Weight vs Horsepower',fontsize = 17)
ax.set_ylabel('Value',fontsize=12)
ax.set_xlabel('Model year',fontsize=12)
ax.legend()
```

```python
import matplotlib.pyplot as plt
import numpy as np
t = np.cos(2 * np.pi * x) * np.exp(-x)
x1 = np.arange(0.0, 5.0, 0.1)
y2 = np.arange(0.0, 5.0, 0.02)
plt.figure()
plt.ylabel('Ordinate',fontsize=12)
plt.xlabel('abcissa',fontsize=12)
plt.annotate('2nd crest', xy=(2, 0.2), xytext=(2.0, 0.2), arrowprops=dict(facecolor='yellow', shrink=.01))
plt.plot(x1, f(x1), 'bo', y2, f(y2), 'k')
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
plt.figure()
plt.ylabel('Ordinate',fontsize=12)
plt.xlabel('abcissa',fontsize=12)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

```
