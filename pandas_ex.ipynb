```python
#.1.Import the data into Python environment as a Pandas DataFrame.
import pandas as pd
df= pd.read_csv('dataset/rainfall.csv',)
df
```

```python
df.info()
```

```python
#.2.Check for missing values, if any and drop the corresponding rows.
df.dropna(inplace = True)
df
```

```python
#.3.Find the district that gets the highest annual rainfall.#.3.
df.groupby('DISTRICT').max().sort_values('ANNUAL',ascending=0)[['ANNUAL']].head(1)
```

```python
#.4.Display the top 5 states that get the highest annual rainfall.
df.groupby('STATE_UT_NAME').sum().sort_values('ANNUAL',ascending=0)[['ANNUAL']]
```

```python
#.5.Drop the columns 'Jan-Feb', 'Mar-May', 'Jun-Sep', 'Oct-Dec'.
df.drop(columns=['Jan-Feb','Mar-May','Jun-Sep','Oct-Dec'],axis=1)
```

```python
df
```

```python
#.6.Display the state-wise mean rainfall for all the months using a pivot table
import numpy as np
pivot=pd.pivot_table(df,index='STATE_UT_NAME',aggfunc=np.mean)
pivot
```

```python
#.7.Display the count of districts in each state.
df.groupby('STATE_UT_NAME').count()[['DISTRICT']]
```

```python
#.8.For each state, display the district that gets the highest rainfall in May. Also display the recorded rainfall.
df.groupby('STATE_UT_NAME').max()[['DISTRICT','MAY']]
```
